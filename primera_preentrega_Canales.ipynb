## Abstract

### Resumen
- En el dataset 
#### Descripción de columnas
- StudentID: Identificador único de estudiantes
#### Detalles demograficos
- Age: La edad de los estudiantes entre los rangos de 15 a 18 años.
- Gender: Genero de los estudiantes, donde 0 representa masculino y 1 representa femenino.
- Ethnicity: La etnia de los estudiantes, codificados como:
    - 0: Caucasico
    - 1: Afroamericano
    - 2: Asiatico
    - 3: Otro
- ParentalEducation: Nivel Educacional de los padres, codificados como:
    - 0: Ninguno
    - 1: High School
    - 2: Some College
    - 3: Bachelor's
    - 4: Higher
#### Hábitos de Estudio
- StudyTimeWeekly: Estudio semanal de tiempo en horas, entre 0 a 20.
- Absences: Numero de ausencias durante el año escolar, rango de 0 a 30.
- Tutoring: Estado de tutorias, donde 0 indica no y 1 si.
- Participación de los padres
- ParentalSupport: Nivel de apoyo de los padres, codificado como:
    - 0: Ninguno
    - 1: Bajo
    - 2: Moderado
    - 3: Alto
    - 4: Muy alto
#### Actividades extracurriculares
- Extracurricular: Participación en actividades extracurriculares, donde 0 indica no y 1 indica si.
- Sports: Participacion en deportes, donde 0 indica no y 1 indica si.
- Music: Participacion en actividades musicales, donde 0 indica no y 1 indica si.
- Volunteering: Participacion en voluntariado, donde 0 indica no y 1 indica si.
#### Rendimiento academico
- GPA: Promedio de calificaciones en una escala de 2.0 a 4.0 , influenciado por los habitos de estudio, participacion de los padres y actividades extracurriculares.
- GradeClass: Clasificacion de los estudiantes basados en su calificaciones de la columna GPA:
    - 0: 'A' (GPA >= 3.5)
    - 1: 'B' (3.0 <= GPA < 3.5)
    - 2: 'C' (2.5 <= GPA < 3.0)
    - 3: 'D' (2.0 <= GPA < 2.5)
    - 4: 'F' (GPA < 2.0)

### Hipótesis

Mediante el estudio de los distintos datos entregados por el dataset, se espera definir puntos claves para el exito academico de los estudiantes, como por ejemplo:
    -Si realiza actividades extracurriculares, estas afectaran de manera positiva al rendimiento académico.
    -Al dedicar mas tiempo al estudio se aumentan las calificaciones de los estudiantes.
    -Mientras mas se ausentan los estudiantes estos disminuyen sus calificaciones.
    -Si el apoyo de los padres se relaciona con las calificaciones y si su nivel educacional se relaciona.

```python
#mapa de calor de la correlacion de columnas
plt.figure(figsize=(18,14))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', vmin=-1,vmax=1)
plt.title('Matriz de Correlación')
plt.show()
```

Se ve por la matriz de correlación que las variables que mas se relacionan son las ausencias con las calificaciones y las ausencias con los promedios de calificaciones.

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Absences', data=df , palette='flare')
plt.title('Ausencias acumuladas por cantidad de estudiantes en el año', fontsize=16)
plt.xlabel('Dias de ausencias', fontsize=12)
plt.ylabel('Cantidad de estudiantes', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Absences', y='GPA', data=df, palette='viridis')
plt.title('Relación entre Ausencias y el promedio de Calificaciones', fontsize=16)
plt.xlabel('Dias de Ausencia', fontsize=12)
plt.ylabel('Promedio de Calificaciones', fontsize=12)
plt.show()
```

Se muestra como los estudiantes al ausentarse mas dias, sus calificaciones van disminuyendo, incluso los outliers siguen esta tendencia.

```python
plt.figure(figsize=(10, 6))
plt.xlabel('Participacion de los padres',fontsize=10)
plt.ylabel('Densidad',fontsize=10)
sns.kdeplot(df['ParentalSupport'], label='Apoyo de los padres', shade=True)
sns.kdeplot(df['ParentalEducation'], label='Educación de los padres', shade=True)


plt.title('Relación de los padres en la educación de los estudiantes', fontsize=16)
plt.legend()
plt.show()
```

Se muestra que aquellos padres en los extremos que tienen mayor o menor educación apoyan menos en los estudios a los estudiantes y aquellos con algun estudio "intermedio" como alguna colegiatura o bachillerato son los que mas apoyan a los estudiantes con sus estudios.

```python
plt.figure(figsize=(10, 6))
plt.xlabel('Actividad',fontsize=10)
plt.ylabel('Cantidad de Estudiantes',fontsize=10)
sns.histplot(df['Sports'], label='Deportes',bins=5)
sns.histplot(df['Music'], label='Musica',bins=6)
sns.histplot(df['Volunteering'], label='Voluntariado',bins=7)
sns.histplot(df['Extracurricular'], label='Extracurricular',bins=8)
plt.title('Actividades Extracurriculares de los estudiantes', fontsize=16)

plt.legend()
plt.show()
```

Cuesta relacionar las calificaciones con las actividades extracurriculares realizadas por los estudiantes.

```python
plt.figure(figsize=(20, 10))
plt.xlabel('Tiempo de Estudio',fontsize=10)
plt.ylabel('GPA',fontsize=10)
sns.regplot(x='GPA',y='StudyTimeWeekly',data=df,
            line_kws={'color':'red'})

plt.title('Participación de los padres en la educación ', fontsize=16)
```

En su mayoria se logra ver que tiende a ir subiendo la calificación al aumentar el tiempo de estudio.
